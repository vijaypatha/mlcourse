# Data Visualization

#### Univariate
- Categorical 
     - There are three main kinds of qualitative data
         - Binary: two mutually exclusive categories: right/wrong, true/false, or accept/reject
         - unordered or nominal: named categories that do not have an implicit or natural value or rank
         - ordered or ordinal: "Short, Medium, or XL
- Quantitative 
    - Continueous: could be divided and reduced to finer and finer levels
    - Discrete: indivisible entities: you can't have 2.5 kids, or 1.3 pets


#### Multivariate 
- Quantitative vs. Quantitative 
- Quantitative vs. Categorical 
- Categorical vs. Categorical

```python
import numpy as np
import pandas as pd 

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()

%config InlineBackend.figure_format = 'svg'
```

```python
data = pd.read_csv('../Data/telecom_churn.csv')
print('Data: Rows and Colms', data.shape)
data.head(5)
```

```python
data.isnull().sum()
```

```python
data.info()
```

### Univariate 

```python
# Categorical: Bar Charts 

sns.countplot(x='Churn', data= data)
```

```python
sns.countplot(x = 'International plan', data=data)
```

```python
sns.countplot('Customer service calls', data=data)
```

### Univariate Quantitiave 

```python
data.columns
```

```python
features = ['Total day minutes','Total intl calls' ]
data[features].hist(figsize=(10,4))
```

```python
data[features].plot(kind = 'density', subplots = True, layout = (1,2), figsize = (10, 4) , sharex = False)
```

```python
sns.boxplot(x='Total intl calls', data=data);
```

### Multivariate Visualization 

**1. Qunatitive vs. Quantitive**

```python
plt.scatter(data['Total day minutes'],data['Total intl calls']);
```

```python
plt.scatter(data['Total day minutes'], data['Total night minutes']);
```

```python
numerical_features = data.select_dtypes(['number'])
```

```python
numerical_features.info()
```

```python
sns.heatmap(numerical_features.corr())
```

**2. Qunatitive vs. Categorical**

```python
sns.lmplot('Total day minutes', 'Total night minutes', data = data, hue='Churn', fit_reg=False)
```

**2. Categorical vs. Categorical**

```python
sns.countplot('Customer service calls', data=data, hue='Churn')
```

```python
sns.countplot('State', data= data, hue='Churn')
```
